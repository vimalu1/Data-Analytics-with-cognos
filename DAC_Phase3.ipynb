```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset =pd.read_csv("Air_quality.csv")
```

```python

# Create subplots for each histogram
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 8))
fig.suptitle("Air Quality Data Histograms")

# Plot histograms for SO2, NO2, RSPM/PM10, and PM 2.5
dataset['SO2'].plot.hist(ax=axes[0, 0], bins=20, edgecolor='k', color='blue')
axes[0, 0].set_title("SO2")

dataset['NO2'].plot.hist(ax=axes[0, 1], bins=20, edgecolor='k', color='green')
axes[0, 1].set_title("NO2")

dataset['RSPM/PM10'].plot.hist(ax=axes[1, 0], bins=20, edgecolor='k', color='red')
axes[1, 0].set_title("RSPM/PM10")

dataset['PM 2.5'].plot.hist(ax=axes[1, 1], bins=20, edgecolor='k', color='purple')
axes[1, 1].set_title("PM 2.5")

# Add labels and show the plots
for ax in axes.flat:
    ax.set_xlabel("Value")
    ax.set_ylabel("Frequency")
    ax.grid(True)

# Adjust spacing between subplots
plt.tight_layout(rect=[0, 0, 1, 0.96])

# Show the plots
plt.show()
```

```python

# Create a box plot using Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(data=dataset[['SO2', 'NO2', 'RSPM/PM10', 'PM 2.5']])
plt.title('Box Plot for Air Quality Data')
plt.xlabel('Pollutants')
plt.ylabel('Values')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

# Show the plot
plt.show()
```

```python
sns.set(style="whitegrid")
sns.jointplot(x='SO2', y='NO2', data=dataset, kind='scatter')

# Show the plot
plt.show()
```

```python
sns.set(style="ticks")
sns.pairplot(dataset, vars=['SO2', 'NO2', 'RSPM/PM10', 'PM 2.5'])

# Show the plot
plt.show()
```

```python
sns.set(style="ticks")
sns.pairplot(dataset, vars=['SO2', 'NO2', 'RSPM/PM10', 'PM 2.5'])

# Show the plot
plt.show()
```

```python
corr_matrix = dataset.corr()

# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap')

# Show the plot
plt.show()
```
